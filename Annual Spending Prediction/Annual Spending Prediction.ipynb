```python
import pandas as pd 
import numpy as np 
import warnings
warnings.filterwarnings('ignore')

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

print(train.shape , test.shape)
print(train.head())
print(test.head())

print(train.isnull().sum())
print(test.isnull().sum())

x_train = train.copy()
x_test = test.copy()

x_train.drop(['Email','Address','Avatar'],axis=1,inplace=True)
print(x_train.shape)

x_test.drop(['Email','Address','Avatar'],axis=1,inplace=True)
print(x_test.shape)

from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()

x_train['Gender'] = label.fit_transform(x_train['Gender'])
x_test['Gender'] = label.fit_transform(x_test['Gender'])

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

y_train = x_train['Yearly Amount Spent']
x_train = x_train.drop('Yearly Amount Spent', axis=1)

x_train = pd.DataFrame(scaler.fit_transform(x_train), columns=x_train.columns)
print(x_train.head())

y_train.reset_index(drop=True, inplace=True)

x_test = pd.DataFrame(scaler.fit_transform(x_test), columns=x_test.columns)
print(x_test.head())

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(x_train, y_train)

y_pred_rf = rf.predict(x_test)

```

```python
sample_predictions = pd.read_csv('test.csv')
sample_predictions['Yearly Amount Spent'] = y_pred_rf
sample_predictions = sample_predictions[['Email','Yearly Amount Spent']]
sample_predictions
```

```python
sample_predictions.to_csv('output.csv',index=False)
```
